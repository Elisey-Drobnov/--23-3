```python
window_size = 2
# Сглаживание с помощью скользящего среднего для всех колонок с использованием min_periods. Циклы Китчина
for column in df_x5_group.columns:
    df_x5_group[column] = df_x5_group[column].rolling(window=window_size, min_periods=1).mean()

for column in df_magnit.columns:
    df_magnit[column] = df_magnit[column].rolling(window=window_size, min_periods=1).mean()
```

```python
def remove_multicollinearity(df, target_variable, threshold=0.7):
    corr_matrix = df.corr().abs()

    to_drop = set()
    
    predictors = [col for col in df.columns if col != target_variable]
    
    for i in range(len(predictors)):
        for j in range(i + 1, len(predictors)):
            colname1 = predictors[i]
            colname2 = predictors[j]
            # Если корреляция между предикторами выше порога
            if abs(corr_matrix.loc[colname1, colname2]) > threshold and abs(corr_matrix.loc[colname1, colname2]) != 1:
                # Сравниваем корреляцию с целевой переменной
                corr_with_target1 = abs(corr_matrix.loc[target_variable, colname1])
                corr_with_target2 = abs(corr_matrix.loc[target_variable, colname2])
                
                # Удаляем переменную с меньшей корреляцией с целевой
                if corr_with_target1 < corr_with_target2:
                    to_drop.add(colname1)
                else:
                    to_drop.add(colname2)

    if to_drop:
        df = df.drop(columns=to_drop)
    
    return df
```

```python
def check_normality(df, df_name):
    # Проверяет нормальность распределения
    results = []
    for column in df.select_dtypes(include=np.number).columns:
        data = df[column].dropna()
        if len(data) < 3:
            continue
        
        stat, p = shapiro(data)
        is_normal = 'Нормальное' if p > 0.05 else 'Не нормальное'
        
        results.append({
            'Компания': df_name,
            'Показатель': column,
            'Статистика теста': f'W = {stat:.3f}',
            'p-value': f'{p:.3f}',
            'Распределение': is_normal
        })
    return pd.DataFrame(results)

results = pd.concat([
    check_normality(df_magnit, 'Magnit'),
    check_normality(df_x5_group, 'X5 Group')
])
```

```python
from IPython.display import display, Math, Latex

data = {
    'EBITDA margin': {
        'beta': [4.819489e+07, 9.204766e+06],
        'se': [8.728741e+06, 2.335394e+06]
    },
    'GMROI': {
        'beta': [2.246508e+06, 1.414378e+08],
        'se': [4.949191e+05, 1.649165e+07]
    }
}

results = pd.DataFrame(
    columns=['Показатель', 'β Магнит', 'SE Магнит', 'β X5', 'SE X5', 'β средневзвешенный'],
    index=range(2)
)


for i, (metric, values) in enumerate(data.items()):
    # Расчет весов
    weights = [1/(se**2) for se in values['se']]
    
    # Расчет средневзвешенного коэффициента
    beta_avg = np.sum([w*b for w,b in zip(weights, values['beta'])]) / sum(weights)
    
    # Запись в таблицу
    results.loc[i] = [
        metric,
        f"{values['beta'][0]:.2e}",
        f"{values['se'][0]:.2e}",
        f"{values['beta'][1]:.2e}",
        f"{values['se'][1]:.2e}",
        f"{beta_avg:.2e}"
    ]

styled_table = (
    results.style
    .set_caption('Результаты расчетов средневзвешенных коэффициентов')
    .set_properties(**{'text-align': 'center'})
    .hide(axis='index')
    .format(precision=2)
)

display(Latex(r'''
1. Расчет весов: 
$$ w_i = \frac{1}{(SE_i)^2} $$

2. Средневзвешенный коэффициент:
$$ \beta_{avg} = \frac{\sum (w_i \cdot \beta_i)}{\sum w_i} $$
'''))

display(styled_table)
```
